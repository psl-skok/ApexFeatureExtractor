# Identifying Person1 and Person2 from call data

```python
import pandas as pd

# 2. define function to assign speakers per NaturalId
def assign_speakers_pdf(pdf):
    last_person1, last_person2, last_speaker = None, None, None
    speakers = []

    for sid_raw in pdf["S"]:
        sid = int(str(sid_raw))
        if last_person1 is None and last_person2 is None:
            speaker = "Person1"
            last_person1 = sid
        elif last_person2 is None:
            if sid != last_person1 + 1:
                speaker = "Person2"
                last_person2 = sid
            else:
                speaker = "Person1"
                last_person1 = sid
        else:
            da = abs(sid - last_person1)
            dc = abs(sid - last_person2)
            if da < dc:
                speaker = "Person1"
                last_person1 = sid
            elif dc < da:
                speaker = "Person2"
                last_person2 = sid
            else:
                speaker = "Person1" if last_speaker == "Person2" else "Person2"

        last_speaker = speaker
        speakers.append(speaker)

    pdf["Speaker"] = speakers
    return pdf

# 3. apply per NaturalId group
speakerIDChart = df.groupby("call_id", group_keys=False).apply(assign_speakers_pdf)

# 4. preview
speakerIDChart.head(265)
```

# Using LLM to determine if Person1 or Person2 is the Agent vs Customer

```python
import os
from dotenv import load_dotenv

load_dotenv()
api_key = os.getenv("OPENAI_API_KEY")

from openai import OpenAI
client = OpenAI(api_key=api_key)

# Function to detect which Person is Agent/Customer for one call
def detect_agent_customer(pdf):
    # Take first ~10 rows for context
    sample_text = "\n".join(
        [f"{row['Speaker']}: {row['Sentence']}" for _, row in pdf.head(10).iterrows()]
    )

    # Prompt for the LLM
    prompt = f"""
We have a conversation transcript with two speakers labeled Person1 and Person2.
Here are the first lines:

{sample_text}

Based on these lines, identify which label is the Agent and which is the Customer.
Respond in the format: Person1=Agent, Person2=Customer (or vice versa).
Do not add anything else.
"""

    # Call OpenAI API
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": prompt}],
        temperature=0,
        max_tokens=30
    )

    # Extract text
    result = response.choices[0].message.content.strip()
    
    # Parse mapping
    mapping = {}
    for part in result.split(","):
        k, v = part.split("=")
        mapping[k.strip()] = v.strip()

    # Apply mapping to entire call
    pdf["Role"] = pdf["Speaker"].map(mapping)
    return pdf

# Apply agent/customer detection to every call
all_calls_labeled = (
    speakerIDChart
    .groupby("call_id", group_keys=False)  # ensures the groupby doesn't insert extra index levels
    .apply(detect_agent_customer)
)

# Drop unnecessary columns after labeling
all_calls_labeled = all_calls_labeled.drop(columns=["Speaker", "Source", "SentenceId"], errors='ignore')


# preview results
all_calls_labeled.head(265)
```

# Condensing full conversation text into one cell with speaker ID

```python
# Collapse each call into one string with line breaks
conversation_summary = (
    all_calls_labeled
    .groupby("call_id")
    .apply(lambda g: " ".join([f"[{row['Role']}] \"{row['Sentence']}\"" for _, row in g.iterrows()]))
    .reset_index(name="call_text")
)

# Show full text in Jupyter/VSCode
pd.set_option("display.max_colwidth", None)
print(conversation_summary.iloc[0]['call_text'])
```

# Grouping Consecutive Sentence

```python
def collapse_consecutive_speakers(df):
    collapsed = []
    last_role = None
    buffer = []

    for _, row in df.iterrows():
        role = row['Role']
        sentence = row['Sentence']

        if role == last_role:
            # Same speaker as before → add to buffer
            buffer.append(sentence)
        else:
            # Speaker changed → flush buffer
            if buffer:
                collapsed.append(f"[{last_role}] \"{' '.join(buffer)}\"")
            buffer = [sentence]
            last_role = role

    # Flush the last buffer
    if buffer:
        collapsed.append(f"[{last_role}] \"{' '.join(buffer)}\"")

    return " ".join(collapsed)


# Apply per call
conversation_summary = (
    all_calls_labeled
    .groupby("call_id")
    .apply(collapse_consecutive_speakers)
    .reset_index(name="call_text")
)

# Show result
pd.set_option("display.max_colwidth", None)
print(conversation_summary.iloc[0]['call_text'])
```
